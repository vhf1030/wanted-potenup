# 4. 배열 변환

```python
# 0에서 10까지의 연속적인 배열
arr1 = np.array([n for n in range(10)])
print(arr1)
```

```python
# reshape
arr2 = arr1.reshape((2,5))
print(arr2)
```

```python
# transpose
arr3 = arr2.transpose()
print(arr3)
print(arr2.T)
```

# 5. 연산

```python
# 사칙연산
arr4 = arr3 * 2
print(arr4)

arr5 = arr4 + 2
print(arr5)
```

```python
arr1 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr1)
print(arr2)

print(arr1 + arr2)
print(arr1 * arr2)

# 행렬 곱셈
print(arr1 @ arr2.T)
print(arr1.T @ arr2)

# 역행렬
print(arr2[:2,:2])
print(np.linalg.inv(arr2[:2,:2]))
```

# 6. 통계

```python
# 기초통계
arr = np.array([40, 50, 60])
print(arr)
print(arr.sum(), arr.min(), arr.max(), np.median(arr), arr.mean(), arr.var(), arr.std())

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
print(arr)
print(arr.shape)
print(arr.sum(), arr.min(), arr.max(), np.median(arr), arr.mean(), arr.var(), arr.std())
print(arr.mean(axis=None))
print(arr.mean(axis=0))
print(arr.mean(axis=1))
```

```python

arr = np.array([
    [
        [23, 4, 88, 100, 98],
        [58, 96, 80, 64, 35],
        [51, 57, 67, 55, 95]
    ], [
        [23, 4, 88, 100, 98],
        [58, 96, 80, 64, 35],
        [51, 57, 67, 55, 95]
    ]
])
print(arr)
print(arr.shape)
print(arr.mean(axis=None))
print(arr.mean(axis=0))
print(arr.mean(axis=1))
print(arr.mean(axis=2))
print(arr.mean(axis=(0,1)))
print(arr.mean(axis=(0,2)))
print(arr.mean(axis=(1,2)))
```
