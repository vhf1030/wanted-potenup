```python
from langchain_huggingface import HuggingFacePipeline

llm = HuggingFacePipeline.from_model_id(
    model_id="Bllossom/llama-3.2-Korean-Bllossom-3B",
    task="text-generation",
    pipeline_kwargs={
        'max_new_token': 300,
        # 'top_k': 50,
        'top_p': 0.5,
        'temperature': 0.1
    }
)

llm.invoke('안녕')
```
