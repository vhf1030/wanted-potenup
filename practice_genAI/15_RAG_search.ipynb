### TravilySearch
- https://app.tavily.com/home

```python
from langchain_community.tools.tavily_search import TavilySearchResults

tool = TavilySearchResults(
    max_results=5,
    include_answer=True,  # 요약된 정보 반환 여부
    include_raw_content=True  # 원본 데이터 반환 여부
)
```

```python
results = tool.invoke({'query': "k-pop 가수 카리나의 국적은?"})
len(results)
```

```python
results
```

```python
from langchain_openai import ChatOpenAI

t_tool = TavilySearchResults(k=3)
llm = ChatOpenAI(tiktoken_model_name='gpt-4o-mini')
tools = [t_tool]

llm_tools = llm.bind_tools(tools=tools)
```

```python
from langchain_core.output_parsers.openai_tools import JsonOutputToolsParser

chain = llm_tools | JsonOutputToolsParser(tools=tools)

result = chain.invoke("카리나가 누구야")
result
```

### DuckDuckGoSearch

```python
from langchain_community.tools import DuckDuckGoSearchRun
search = DuckDuckGoSearchRun(k=3)

print(search.invoke("카리나가 누구야"))

t_tool = DuckDuckGoSearchRun(k=3)
llm = ChatOpenAI(tiktoken_model_name='gpt-4o-mini')
tools = [t_tool]

llm_tools = llm.bind_tools(tools=tools)
chain = llm_tools | JsonOutputToolsParser(tools=tools)

result = chain.invoke("카리나가 누구야")
result
```

```python
from langchain.agents.initialize import initialize_agent, AgentType

ddg_tool = DuckDuckGoSearchRun(k=3)
youtube_tool = YouTubeSearchTool(k=3)
llm = ChatOpenAI(tiktoken_model_name='gpt-4o-mini')
tools = [ddg_tool, youtube_tool]

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)

result = agent('토토로 주제곡 찾아서 영상 링크 알려줘')
result
```
