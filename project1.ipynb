### 목표: 여행객들을 대상으로 어떤 관광상품을 개발하는 것이 좋을까?

변수
- nat: 국가
- expense: 비용
- bestact: 여행목적

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import koreanize_matplotlib

ftour = pd.read_csv('data/ftour.csv', encoding='cp949')

ftour.head()

print(ftour[:10])
```

```python
print(ftour.describe())
```

```python
# ftour[['nat', 'expense']].groupby('nat').plot(kind='box')
print(ftour['expense'])
ftour.boxplot(column=['expense'], by='nat')
plt.xticks(rotation=45)
plt.yscale('log')
plt.ylabel('비용(log scale)')
plt.xlabel('국가')
plt.show()
```

```python
ftour.boxplot(column=['expense'], by='member')
plt.xticks(rotation=45)
plt.yscale('log')
plt.ylabel('비용(log scale)')
plt.xlabel('멤버')
plt.show()
```

```python
ftour[ftour['id'] == 840]
```

```python
bp = ftour.boxplot(
    column=['expense'],
    by='nat',
    showmeans=True,
    meanprops={
        'markeredgecolor': 'firebrick',
        'markerfacecolor': 'firebrick'
    }
)
plt.yscale('log')
plt.show()
```

```python
# for col in ftour.columns:
columns = [
    'job', 'nat', 'decision', 'object', 'mov1', 'mov2', 'mov3',
    'act1', 'act2', 'act3', 'bestact', 'accom', 'shop1', 'shop2', 'shop3'
]

for col in columns:
    bp = ftour.boxplot(
        column=['expense'],
        by=col,
        showmeans=True,
        meanprops={
            'markeredgecolor': 'firebrick',
            'markerfacecolor': 'firebrick'
        }
    )
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.ylabel('비용(log scale)')
    plt.xlabel(col)
    plt.show()


```
