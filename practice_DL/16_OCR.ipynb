### OCR
- 이미지에서 텍스트 추출
- 목표: 웹툰의 한글을 영어로 번역

#### 이미지 불러오기

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "./data/images/webtoon1.jpeg"
image = Image.open(image_path).convert("RGB")  # RGB로 변환

plt.figure(figsize=(3,3))
plt.imshow(image)
plt.axis("off")
plt.show()
```

#### 이미지에서 텍스트 추출하기

```python
from paddleocr import PaddleOCR
ocr = PaddleOCR(lang="korean")

result = ocr.ocr(image_path, cls=False)
```

```python
print(result)
print(len(result))
```

```python
for x in result[0]:
    print(len(x))
    
    print(x)
```

```python
# 미션
# 빈 리스트를 만든다 - box_data, text_data, score_data
# 11개의 데이터를 좌표별, 텍스트별, 신뢰도별로 담는다.

# box_data = [r[0] for r in result[0]]
# text_data = [r[1][0] for r in result[0]]
# score_data = [r[1][1] for r in result[0]]
box_data, text_data, score_data = zip(*[(r[0], r[1][0], r[1][1]) for r in result[0]])

print(box_data)
print(text_data)
print(score_data)
```

```python
# 기존의 이미지에 박스 그리기

from paddleocr import draw_ocr
font_path = "C:/Windows/Fonts/malgun.ttf"
boxes_np = draw_ocr(image, box_data, text_data, score_data, font_path=font_path)
print(boxes_np[0])

plt.figure(figsize=(10,10))
plt.imshow(boxes_np)
plt.axis("off")
plt.show()
```
