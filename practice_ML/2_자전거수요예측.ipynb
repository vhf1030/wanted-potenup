```python
import pandas as pd
```

```python
bike_train = pd.read_csv("train.csv")
bike_train.head()
```

```python
bike_test = pd.read_csv("test.csv")
bike_test.head()
```

```python
bike_train.info()
```

```python
bike_train['datetime'] = bike_train.datetime.apply(pd.to_datetime)
bike_train.info()
```

```python
bike_train.describe()
```

```python
# 년, 월, 일, 시간 추출
bike_train["year"] = bike_train["datetime"].apply(lambda x: x.year)
bike_train["month"] = bike_train["datetime"].apply(lambda x: x.month)
bike_train["day"] = bike_train["datetime"].apply(lambda x: x.day)
bike_train["hour"] = bike_train["datetime"].apply(lambda x: x.hour)
bike_train.head()

# 요일 추출
bike_train["day_of_week"] = bike_train["datetime"].apply(lambda x: x.day_of_week)
```

```python
bike_train = bike_train.drop(["datetime", "casual", "registered"], axis=1)
bike_train.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.subplot(2,3,1)
sns.barplot(x="season", y="count", data=bike_train)
plt.subplot(2,3,2)
sns.histplot(data=bike_train["count"], kde=True, linewidth=0)
plt.subplot(2,3,3)
plt.subplot(2,3,4)
plt.subplot(2,3,5)
plt.subplot(2,3,6)
plt.show()

```

```python
# 선형회귀와 다항회귀 비교
x_scatter = bike_train[['hour']].values
y_scatter = bike_train['count'].values
x_scatter
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# 1. 선형회귀 모델 선택
lr = LinearRegression()

# 2. 모델 학습
lr.fit(X=x_scatter, y=y_scatter)

# 3. 예측하기
pred = lr.predict(x_scatter)
print(pred)
# 4. 시각화
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter)  # 실제 데이터 그리기
plt.plot(x_scatter, pred, color="red")  # 예측한 모델 그리기
plt.show()

```

```python
# 1. 다항회귀 모델 선택
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x_scatter)
print(x_poly)

# 2. 모델 학습
poly_model = LinearRegression()
poly_model.fit(X=x_poly, y=y_scatter)

# 3. 예측하기
poly_pred = poly_model.predict(x_poly)
print(poly_pred)

# 4. 시각화
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter)  # 실제 데이터 그리기
plt.plot(x_scatter, poly_pred, color="red")  # 예측한 모델 그리기
plt.show()
```

```python
bike_train.head()
X = bike_train.drop(["count"], axis=1)
y = bike_train["count"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# 모델 학습
lr = LinearRegression()
lr.fit(X_train, y_train)

```

```python
from sklearn.metrics import *

# 모델 평가하기
pred = lr.predict(X_test)

mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)

print("mse", mse)
print("r2", r2)
```

```python
X_dummies = pd.get_dummies(X, columns=['season', 'weather'])
X_dummies.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_dummies, y, test_size=0.2, random_state=11)

# 모델 학습
lr = LinearRegression()
lr.fit(X_train, y_train)

# 모델 평가하기
pred = lr.predict(X_test)

mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)

print("mse", mse)
print("r2", r2)
```
