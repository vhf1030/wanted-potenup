```python
# 데이터 3000개 가져오기
# 데이터는 한 번 요청할 때 1,000개를 넘을 수 없다
# 1, 1000
# 1001, 2000
# 2001, 3000
result = []
for i in range(3):
    start = i * 1000 + 1
    end = (i+1) * 1000
    print(start, end)
    url = f"http://openapi.seoul.go.kr:8088/{api_key}/json/tbCycleRentUseMonthInfo/{start}/{end}/202208/"
    print(url)
    response = requests.get(url)
    data = response.json()
    print(data)
    result.extend(data['cycleRentUseMonthInfo']['row'])

print(len(result))
```
