```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

df_credit = pd.read_csv("../../data/creditcard.csv")
df_credit.head()
```

```python
df_credit.describe()
```

```python
from sklearn.model_selection import train_test_split

# train, test 셋 분리
X = df_credit.drop("Class", axis=1)
y = df_credit["Class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=11)
y_train.value_counts()
```

```python
# 각 주성분의 분산 확인
variances = X_train.loc[:,"V1":"V28"].var(axis=0)  # PCA 결과값의 분산 계산

# 분산 출력
print("주성분별 분산:")
print(variances)

# 주성분별 분산 시각화
plt.figure(figsize=(10, 6))
plt.bar(range(len(variances)), variances, color='skyblue')
plt.title("PCA 주성분별 분산", fontsize=16)
plt.xlabel("주성분 번호", fontsize=12)
plt.ylabel("분산 크기", fontsize=12)
plt.xticks(ticks=range(len(variances)), labels=list(variances.index))
plt.show()
```
