### YOLO 실습

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "../data/images/yolo_image.jpg"
image = Image.open(image_path).convert("RGB")  # RGB로 변환
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
model

```

```python
print(model.names)
results = model(image)
```

```python
# print(results[0])
# print(dir(results[0]))
print(results[0].boxes)
```

```python
result_cls = results[0].boxes.cls
print(result_cls)
```

```python
result_conf = results[0].boxes.conf
print(result_conf)
```

```python
result_box = results[0].boxes.xyxy
print(result_box)
```

```python
# 그림 그리기
import cv2
import numpy as np

image_np = np.array(image)

for idx, box in enumerate(result_box):
    x1, y1, x2, y2 = (int(b) for b in box)
    # print(x1, y1, x2, y2)
    
    cv2.rectangle(image_np, (x1, y1), (x2, y2), (0, 255, 0), 3)
    
    # 신뢰도 점수 표시
    cls_idx = int(result_cls[idx].item())
    cls_name = model.names[cls_idx]
    
    score = result_conf[idx].item() * 100  # 0~1 값을 100%로 변환
    cv2.putText(
        image_np, f"{cls_name} {score:.1f}%", (x1, y2 + 15),
        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2
    )


plt.figure(figsize=(10, 10))
plt.imshow(image_np)
plt.axis("off")
plt.show()
    
```
