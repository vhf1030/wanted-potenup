```python
import requests
url = "http://localhost:8080/basic/all"
response = requests.get(url)
response.json()
```

```python
import requests
url = "http://localhost:8080/basic/query"
params = {"category": "Electronics"}
response = requests.get(url, params=params)
response.json()
```

```python
import requests

url = "http://localhost:8080/basic/body/text"
data = {
    "name": "Mouse",
    "category": "Electronics",
    "price": 2000
}

response = requests.post(url, json=data)
response.json()
```

```python
import requests

image_path = "./test.jpg"
with open(image_path, "rb") as f:
    image_data = f.read()

print(image_data)

url = "http://localhost:8080/basic/body/image/binary"

data = {
    "file": image_data
}

response = requests.post(url, files=data)
response.json()
```

```python
import requests

url = "http://localhost:8080/basic/body/image/file"

image_path = "./celeb_test.jpg"
with open(image_path, "rb") as f:
    files = {'file': ("celeb_test.jpg", f, "image/jpeg")}
    response = requests.post(url, files=files)

response.json()
```
